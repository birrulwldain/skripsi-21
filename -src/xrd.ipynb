```python
from pymatgen.io.cif import CifParser
from pymatgen.analysis.diffraction.xrd import XRDCalculator
import matplotlib.pyplot as plt

# Baca file CIF
parser = CifParser("a.cif")
structure = parser.get_structures()[0]

# Simulasi XRD
xrd_calculator = XRDCalculator(wavelength="CuKa")
xrd_pattern = xrd_calculator.get_pattern(structure)

# Plot spektrum XRD
plt.plot(xrd_pattern.x, xrd_pattern.y, label="Hydroxyapatite")
plt.xlabel("$2\\theta$ (degrees)")
plt.ylabel("Intensity")
plt.title("Simulated XRD Pattern")
plt.legend()
plt.show()
```

```python
# Buka file CIF
with open("mp-41472.cif", "r") as file:
    cif_data = file.readlines()

# Perbaiki nilai fraksional
fixed_cif_data = []
for line in cif_data:
    line = line.replace("1/3", "0.3333").replace("2/3", "0.6667")
    fixed_cif_data.append(line)

# Simpan file CIF yang telah diperbaiki
with open("mp-41472_fixed.cif", "w") as file:
    file.writelines(fixed_cif_data)

print("File CIF berhasil diperbaiki dan disimpan sebagai mp-41472_fixed.cif")
```
